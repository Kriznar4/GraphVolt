```python

def add_missing_measurements(df_measurements, node_ids_list, mode):
    """
    Add missing measurements to df_measurements for each node_id.
    Different modes avaliable:
    - "mean": add mean of all measurements for each node_id except _power is set to 0
    """
    #get mean timeseries of all measurements
    mean_alltime = df_measurements.groupby("date_time").agg("mean").reset_index()
    #set active_power and reactive_power to 0
    mean_alltime["active_power"] = 0
    mean_alltime["reactive_power"] = 0

    #find missing nodes
    missing_nodes = [i for i in node_ids_list if i not in df_measurements["node_id"].unique()]

    for node in missing_nodes:
        print(f'missing node {node} filling in')
        mean_alltime["node_id"] = node
        df_measurements = pd.concat([df_measurements, mean_alltime], ignore_index=True)

    return df_measurements

class GraphVoltDatasetLoader(object):
    '''
        Check this https://pytorch-geometric-temporal.readthedocs.io/en/latest/_modules/torch_geometric_temporal/dataset/wikimath.html#WikiMathsDatasetLoader
        for an example of how to implement a dataset loader

        And here are the docs https://pytorch-geometric-temporal.readthedocs.io/en/latest/modules/signal.html
    '''
    def __init__(self, trafo_id, lags):
        self._trafo_id = trafo_id
        #number of historic timesteaps to consider : lags
        self._lags  = lags
        self._read_data()
    
    def _read_data(self):
        self._dataset = preprocess(read_raw_network_data(trafo_id, depth=1)[0])
        self._dataset["time_periods"] = len(self._dataset["measurements"]["date_time"].unique())

    def _get_edges(self):
        self._edges = self._dataset["edges_static_data"][["from_node_id", "to_node_id"]].values.T
    
    def _get_edge_weights(self):
        self._edges_weights = self._dataset["edges_static_data"].drop(["from_node_id", "to_node_id", "coords", "coords_transformed", "type"], axis=1).values

    def _get_targets_and_features(self):
        self._dataset["measurements"] = add_missing_measurements(self._dataset["measurements"], list(self._dataset["nodes_static_data"]["node_id"]), mode="mean")

        #drop unnnecesary columns
        df_static = self._dataset["nodes_static_data"].drop(columns=["smm_list", "prikljucna_moc_odjem", "prikljucna_moc_oddaja", "x_y", "lon_lat"])

        #if aclass_id is "PMO" or "TR" set it to 1 else 0
        df_static["aclass_id"] = df_static["aclass_id"].apply(lambda x: 1 if x in ["PMO", "TR"] else 0)

        #fill NaN with 0 in prikljucna_moc_odjem_aggr and prikljucna_moc_oddaja_aggr
        df_static["prikljucna_moc_odjem_aggr"] = df_static["prikljucna_moc_odjem_aggr"].fillna(0)
        df_static["prikljucna_moc_oddaja_aggr"] = df_static["prikljucna_moc_oddaja_aggr"].fillna(0)

        #add static data to each row
        self._dataset["measurements"] = self._dataset["measurements"].merge(df_static, on="node_id", how="left")

        #add columns minute, hour, day, month, year to self._dataset["measurements"]
        self._dataset["measurements"]["minute"] = self._dataset["measurements"]["date_time"].dt.minute
        self._dataset["measurements"]["hour"] = self._dataset["measurements"]["date_time"].dt.hour
        self._dataset["measurements"]["day"] = self._dataset["measurements"]["date_time"].dt.day
        self._dataset["measurements"]["month"] = self._dataset["measurements"]["date_time"].dt.month
        self._dataset["measurements"]["year"] = self._dataset["measurements"]["date_time"].dt.year

        #create features and targets

        #get sorted series of time_steps
        time_steps = self._dataset["measurements"].sort_values("date_time")["date_time"].unique()

        #get index of voltage column
        some_df = self._dataset["measurements"].groupby("date_time").get_group(self._dataset["measurments"]["date_time"][0]).sort_values("node_id").drop(["date_time", "node_id"], axis=1)
        voltage_idx = some_df.columns.get_loc("voltage") 

        #get list of np.arrays (2D) of features for each time_step    
        features = [features.append(self._dataset["measurements"].groupby("date_time").get_group(date_time).sort_values("node_id").drop(["date_time", "node_id"], axis=1).values) for date_time in time_steps]
        stacked_features = np.stack(features)

        #normalize TODO
        standardized_features = stacked_features

        #get list of lags for each time_step
        self.features = [
            standardized_features[i : (i + self._lags), : , voltage_idx ]
            for i in range(len(standardized_features) - self._lags)
        ]

        self.targets = [
            standardized_features[i + self._lags, : , : ]
            for i in range(len(standardized_features) - self._lags)
        ]

    

    def get_dataset(self):
        self._get_edges()
        self._get_edge_weights()
        self._get_targets_and_features()
        dataset = StaticGraphTemporalSignal(
            self._edges, self._edge_weights, self.features, self.targets
        )
        return dataset
```
